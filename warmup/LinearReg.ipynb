```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
D = 2
N = 1000

X1 = np.linspace(start=0, stop=5, num=N)
X2 = X1 + np.random.randn(N) * 0.1
Y = X1.reshape(N, 1)
X = np.stack((X1, X2), axis = 1)
plt.scatter(X[:, 0], X[:, 1])
```

```python
W = np.random.randn(D, 1) / np.sqrt(D)
plt.scatter(X[:, 0], X[:, 1], s = 5)
plt.plot(X1, X.dot(W), c='red')
plt.title('Before training')
```

```python
#using gradient of mse 
#W = np.random.randn(D, 1) / np.sqrt(D)
costs = []
for i in range(1000) :
    W -= 0.01 * (2 * X.T.dot(X.dot(W) - Y)) / N
    cost = (Y - X.dot(W)).sum() / N
    costs.append(cost)
    if i % 20 == 0:
        print(f'Epoch {i} : Cost {cost}')
```

```python
plt.plot(costs)
```
