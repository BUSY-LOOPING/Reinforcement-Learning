```python
import numpy as np
import os
os.chdir(os.path.join('..', 'dynamic_programming'))
import import_ipynb
from grid_world import windy_grid, ACTION_SPACE
```

```python
grid = windy_grid()
all_states = grid.actions.keys()
grid.set_state(list(all_states) [np.random.choice(len(all_states))])
policy = {}
for s in grid.actions.keys() :
    policy[s] = np.random.choice(ACTION_SPACE)
s = grid.current_state()
states = [s]
rewards = []
count = 0

while not grid.game_over() and count < 200:
    count += 1
    a = policy[s]
    r = grid.move(a)
    rewards.append(r)
    s = grid.current_state()
    states.append(s)
```
